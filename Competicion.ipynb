```

from sklearn.base import TransformerMixin
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.feature_selection import SelectKBest, SelectPercentile, chi2, f_classif, f_regression

class ValExtractor():
    def __init__(self,number_attributes=10,dic_trees={}):
        self.number_attributes=number_attributes
        self.dic=dic_trees
        self.mean=[]
        self.standar_dev=[]
        
    def transform(self,X,*_):
        X_to_transform=X
        for i in self.dic:
            #print(i)
            try:
                rows_to_be_fixed_2=np.arange(X_to_transform.shape[0])[X_to_transform.iloc[::,i].isna()]
                #print(rows_to_be_fixed_2)
                X_to_transform.iloc[rows_to_be_fixed_2,i]=self.dic[i][0].predict(X_to_transform.iloc[rows_to_be_fixed_2,self.dic[i][1]])
            except:
                pass
        return X_to_transform
            #print(rows_to_be_fixed)
    
    def fit_transform(self, X_temp,*_):
        
        X_train_cl=X_temp
        features_without_NaN=(X_train_cl.isna().sum()==0)
        
        features_with_NaN=np.arange(len(features_without_NaN))[np.logical_not(features_without_NaN)]
        features_without_NaN=np.arange(len(features_without_NaN))[features_without_NaN]

        features_with_NaN_dict={}

        for i in features_with_NaN:
            rows_to_be_fixed=np.arange(X_train_cl.shape[0])[X_train_cl.iloc[::,i].isna()]
            rows_used_to_fix=np.arange(X_train_cl.shape[0])[np.logical_not(X_train_cl.iloc[::,i].isna())]
            features_with_NaN_dict[i]=[rows_used_to_fix,
                                       rows_to_be_fixed]
        for i in features_with_NaN_dict:
            kbest=SelectKBest(score_func=f_regression).fit(X=X_train_cl.iloc[features_with_NaN_dict[i][0],features_without_NaN],y=X_train.iloc[features_with_NaN_dict[i][0],i])
            kbest_features=np.argsort(-kbest.scores_)[0:self.number_attributes]
            features_with_NaN_dict[i]=[features_with_NaN_dict[i][0],
                                       features_with_NaN_dict[i][1],
                                       features_without_NaN[kbest_features]]
        
        dic_trees={}
        for i in features_with_NaN_dict:
            regr = RandomForestRegressor(max_depth=8, random_state=0, n_estimators=50)
            regr.fit(X=X_train_cl.iloc[features_with_NaN_dict[i][0],
                                       features_with_NaN_dict[i][2]],
                     y=X_train_cl.iloc[features_with_NaN_dict[i][0],i])
            
            X_train_cl.iloc[features_with_NaN_dict[i][1],i]=regr.predict(X=X_train_cl.iloc[features_with_NaN_dict[i][1],features_with_NaN_dict[i][2]])
            dic_trees[i]=[regr,features_with_NaN_dict[i][2]]
        self.dic=dic_trees
        return X_train_cl
```
